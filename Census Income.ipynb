```python
#Need to LAbel encode ordinal data ('workclass' 'education','marital.status', 'occupation', 'relationship', 'race', 'sex', 'native.country','income')
```

```python
from sklearn.preprocessing import LabelEncoder
le1=LabelEncoder()
le3=LabelEncoder()
le5=LabelEncoder()
le6=LabelEncoder()
le7=LabelEncoder()
le8=LabelEncoder()
le9=LabelEncoder()
le13=LabelEncoder()
le=LabelEncoder()
x['workclass']=le1.fit_transform(x['workclass'])
x['education']=le3.fit_transform(x['education'])
x['marital.status']=le5.fit_transform(x['marital.status'])
x['occupation']=le6.fit_transform(x['occupation'])
x['relationship']=le7.fit_transform(x['relationship'])
x['race']=le8.fit_transform(x['race'])
x['sex']=le9.fit_transform(x["sex"])
x['native.country']=le13.fit_transform(x["native.country"])
y=le.fit_transform(y)
```

```python
#Splitting Dataset into Training set and Test set
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
#!pip install xgboost
from xgboost import XGBClassifier
classifier=XGBClassifier()
classifier.fit(X_train,Y_train)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
y_pred=classifier.predict(X_test)
cm=confusion_matrix(Y_test,y_pred)
print(cm)
accuracy_score(Y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,y_pred))
```
